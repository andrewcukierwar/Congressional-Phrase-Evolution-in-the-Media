```python
procedural = set()
file = open("vocabulary/procedural.txt", 'r')
next(file)
for line in file.readlines():
    phrase = line[:line.find('|')] if '|' in line else line
    word1, word2 = phrase.split(" ")
    bigram = (word1, word2)
    procedural.add(bigram)
file.close()
```
